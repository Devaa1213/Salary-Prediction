```python
#We imported pandas to ma
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("/content/Salary_Data.csv")
data
```

```python
data
x = pd.DataFrame(data["YearsExperience"])
y = data["Salary"]
```

```python
plt.scatter(x,y)
```

```python
model = LinearRegression()
```

```python
model.fit(pd.DataFrame(x),y)
```

```python
model.predict([[6.2]])
```

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x))
```

```python
c = model.intercept_
c
```

```python
m = model.coef_
m
```

```python
y = m * 6.2 + c
```
